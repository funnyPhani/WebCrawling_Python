```python
try:
    ti = []
    au = []
    li = []
    import requests
    from bs4 import BeautifulSoup
    for pg  in range(1,4):
        
        url = f"https://quotes.toscrape.com/page/{pg}"
        print(url)
        req = requests.get(url)
        # print(req.content)
        soup = BeautifulSoup(req.content,"html.parser")
        # print(soup.prettify())
        for i in soup.findAll("div",{"class":"quote"}):
            for j in i.findAll("span",{"class":"text"}):
                ti.append(j.text)
            for k in i.findAll("small",{"class":"author"}):
                au.append(k.text)
            for l in i.findAll("a"):
                li.append(url+l.get("href"))
                break
except Expection as e:
    print(e)
```

```python
import pandas as pd
df = pd.DataFrame([ti,au,li]).T
df.columns = ["quote","Author","url"]
df
```

```python
df.to_excel("quotesData.xlsx",index = False)
```

```python

def getQuoteData(pages):
    try:
        Z = {}
        import requests
        from bs4 import BeautifulSoup
        for pg  in range(1,pages+1):

            url = f"https://quotes.toscrape.com/page/{pg}"
            print(url)
            req = requests.get(url)
            # print(req.content)
            soup = BeautifulSoup(req.content,"html.parser")
            D = []
            # print(soup.prettify())
            for i in soup.findAll("div",{"class":"quote"}):
                d = {}
                for j in i.findAll("span",{"class":"text"}):
                    d['quote']= j.text
                    # ti.append(j.text)
                for k in i.findAll("small",{"class":"author"}):
                    # au.append(k.text)
                    d['author'] = k.text
                for l in i.findAll("a"):
                    # li.append(url+l.get("href"))
                    d['link'] = url+l.get("href")
                    D.append(d)
                    Z[f"page_{pg}"] = D
                    break
        
        return Z
    except Expection as e:
        print(e)
```
